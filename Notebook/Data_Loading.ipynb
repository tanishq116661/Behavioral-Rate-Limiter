#### Loading Dataset and perform some initial preprocessing


**Dataset Used**: CIC-IDS2017
* The CIC-IDS2017 dataset is a comprehensive collection of network traffic data that includes both benign and malicious activities. It is widely used for research in intrusion detection systems (IDS) and cybersecurity.

* The dataset contains various features extracted from network traffic, such as flow duration, total bytes, and packet counts, which can be used to train machine learning models for detecting cyber threats.

* The dataset is downloaded from Kaggle: [CIC-IDS2017 Dataset](https://www.kaggle.com/datasets/dhoogla/cicids2017)

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("dhoogla/cicids2017")

print("Path to dataset files:", path)
```

```python
import os

os.listdir(path)
```

```python
!uv pip install pyarrow fastparquet
```

```python
import pandas as pd 

files = os.listdir(path)

all_df = []

for file in files:
    print(f"Reading {file}...")
    df = pd.read_parquet(os.path.join(path, file))
    
    df.columns = df.columns.str.strip()
    
    all_df.append(df)
```

```python
all_df[0].head()
```

```python
import numpy as np

full_dataset = pd.concat(all_df, ignore_index=True)

full_dataset.replace([np.inf, -np.inf], np.nan, inplace=True)
full_dataset.dropna(inplace=True)
```
